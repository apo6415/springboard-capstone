```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

red_location = r'./winequality-red.csv'
white_location = r'./winequality-white.csv'

redWineDataframe = pd.read_csv(red_location, sep=';')

redWineDataframe.columns = redWineDataframe.columns.str.strip().str.replace('"', '') 

whiteWineDataframe = pd.read_csv(white_location, sep=';')
whiteWineDataframe.columns = whiteWineDataframe.columns.str.strip().str.replace('"', '')

# plt.copper()
red_corr_matrix = redWineDataframe.corr()
white_corr_matrix = whiteWineDataframe.corr()
# corr_plot = plt.matshow(redWineDataframe.corr())
# plt.colorbar(mappable=corr_plot)

fig, ax = plt.subplots(nrows=1, ncols=1)
# plot the heatmap
red = sns.heatmap(red_corr_matrix, 
        xticklabels=red_corr_matrix.columns,
        yticklabels=red_corr_matrix.columns)
fig1,ax1 = plt.subplots(nrows=1, ncols=1)
white = sns.heatmap(white_corr_matrix, 
        xticklabels=white_corr_matrix.columns,
        yticklabels=white_corr_matrix.columns)


# hist_corr_matrix = pd.plotting.scatter_matrix(
#     redWineDataframe, grid=True,diagonal='hist', figsize=(100,100),
#     hist_kwds={}
# )
```
