```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices

only_feature_red_wine = redWineDataframe
only_feature_red_wine.drop('quality', 'columns')
only_feature_red_wine.drop('alcohol', 'columns')
# get y and X dataframes based on this regression:
y_for_VIF, X_for_VIF = dmatrices('binary_quality_index ~ fixed_acidity  + citric_acid + volatile_acidity + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + alcohol + pH + sulphates', only_feature_red_wine, return_type='dataframe')

normalized_X = RobustScaler().fit_transform(X_for_VIF)

# For each X, calculate VIF and save in dataframe
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(normalized_X, i) for i in range(X_for_VIF.shape[1])]
vif["features"] = X_for_VIF.columns



vif.round(1)

print(vif)

plot_vif = plt.plot(vif["features"], vif["VIF Factor"].values.tolist(), label='VIF factor vs Feature')
plt.title('VIF factor vs Feature')
plt.xticks(vif["features"], vif["features"], rotation='vertical')
plt.axvline(x='fixed_acidity', linestyle='dotted')
plt.axvline(x='density', linestyle='dotted')
plt.axhline(y=5.711998, linestyle='dotted')
plt.axhline(y=5.250172, linestyle='dotted')
plt.show(plot_vif)
```

```python
only_feature_white_wine = whiteWineDataframe
only_feature_white_wine.drop('quality', 'columns')
only_feature_white_wine.drop('alcohol', 'columns')
# get y and X dataframes based on this regression:
y_for_VIF, X_for_VIF = dmatrices('binary_quality_index ~ fixed_acidity  + citric_acid + volatile_acidity + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + alcohol + density + pH + sulphates', only_feature_white_wine, return_type='dataframe')

normalized_X = RobustScaler().fit_transform(X_for_VIF)

# For each X, calculate VIF and save in dataframe
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(normalized_X, i) for i in range(X_for_VIF.shape[1])]
vif["features"] = X_for_VIF.columns
print(vif)


vif.round(1)
plot_vif = plt.plot(vif["features"], vif["VIF Factor"].values.tolist(), label='VIF factor vs Feature')
plt.title('VIF factor vs Feature')
plt.xticks(vif["features"], vif["features"], rotation='vertical')
plt.axvline(x='residual_sugar', linestyle='dotted')
plt.axvline(x='density', linestyle='dotted')
plt.axhline(y=11.910102, linestyle='dotted')
plt.axhline(y=26.669135, linestyle='dotted')
plt.show(plot_vif)
```
